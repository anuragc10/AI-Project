```python
import tensorflow as tf
import pathlib
import random
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/ai project/ai/FETAL_PLANES_DB_data.csv')
df_train.head()
```

```python
df_train['Image_name'] = df_train['Image_name;Patient_num;Plane;Brain_plane;Operator;US_Machine;Train '].str.split(';')
```

```python
df_train.head()
```

```python
df_models = pd.DataFrame(df_train['Image_name'].tolist()).fillna('').add_prefix('model_')
```

```python
df = pd.concat([df_train, df_models], axis=1)
df.head()
```

```python
df.drop(['Image_name;Patient_num;Plane;Brain_plane;Operator;US_Machine;Train '], axis=1)
df.drop(['Image_name'], axis=1)
df.head(50)
```

```python
from sklearn.model_selection import train_test_split

training_data, testing_data = train_test_split(df, test_size=0.2, random_state=25)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")


```

```python
IMAGE_SIZE = (256, 256, 1)

data_dir = pathlib.Path('/content/drive/MyDrive/ai project/ai/Images')

```

```python
image = img.imread(str(data_dir)+'/'+(df.iloc[random.randint(0, len(df_train)),2])+'.png')
plt.imshow(image)
```

```python
training_data['model_0'] = training_data['model_0'].apply(lambda x : x + '.png' )
testing_data['model_0'] = testing_data['model_0'].apply(lambda x : x + '.png' )
```

```python
from keras_preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale = 1./255, validation_split=0.2,shear_range=0.2,rotation_range=20,brightness_range=[0.4,1.5])
```

```python
train_ds = datagen.flow_from_dataframe(dataframe=training_data,
                                       directory=data_dir,
                                       x_col = 'model_0',
                                       y_col = 'model_6',
                                       subset='training',
                                       class_mode='categorical',
                                       color_mode='rgb',
                                       seed=43,
                                       target_size=IMAGE_SIZE[:-1],
                                       batch_size=32)

valid_ds = datagen.flow_from_dataframe(dataframe=training_data,
                                       directory=data_dir,
                                       x_col = 'model_0',
                                       y_col = 'model_6',
                                       subset='validation',
                                       class_mode='categorical',
                                       color_mode='rgb',
                                       seed=43,
                                       target_size=IMAGE_SIZE[:-1],
                                       batch_size=32)
test_ds = datagen.flow_from_dataframe(dataframe=testing_data,
                                       directory=data_dir,
                                       x_col = 'model_0',
                                       class_mode=None,
                                       color_mode='rgb',
                                       target_size=IMAGE_SIZE[:-1],
                                       batch_size=32)
```

```python
import tensorflow.keras.backend as K
```

```python
def get_f1(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
from tensorflow.keras.optimizers import Adam 
```

```python
model = tf.keras.models.Sequential([
        
        tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(256, 256, 3)),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(2, activation='softmax')
        
])
```

```python
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=[get_f1])
history = model.fit(train_ds, epochs=1, batch_size=32, shuffle=True, validation_data=valid_ds)
```

```python
prediction = model.predict(test_ds)
```

```python
testing_data['model_6']
```

```python
tp = 0
tn = 0
fp = 0
fn = 0
expected = testing_data['model_6'].values
for i in range(0, len(prediction)):
  pred = 1
  if prediction[i][0] > prediction[i][1]:
    pred = 0
  if pred == 1 and int(expected[i]) == 1:
    tp += 1
  elif pred == 1 and int(expected[i]) == 0:
    fp += 1
  elif pred == 0 and int(expected[i]) == 0:
    tn += 1
  else:
    fn += 1

print(tp, tn, fp, fn)
```

```python
acc = (tp + tn) / (tp + tn + fp + fn)
sens = tp / (tp + fn)
spec = 1 - sens
p = tp / (tp + fp)
r = tp / (tp + fn)
f1 = 2 * p * r / (p + r)
```

```python
print('Accuracy is: ', acc)
print('Recall is: ', r)
print('Precision is: ', p)
print('F1 Score is: ', f1)
print('Sensitivity is: ', sens)
print('Specificty is: ', spec)
```
